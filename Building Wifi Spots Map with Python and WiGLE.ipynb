```python
import requests
import pandas as pd
import numpy as np
import folium
```

### add your encoded token here

```python
token = "your encoded token from wigle.net goes here "
```

```python
header = str("Basic %s" % token)
```

### Lat-Long Coordinates 

```python
latrange1 = 12.7585 
latrange2 = 13.1105 
longrange1 = 77.5477 
longrange2 = 77.8431
```

### Building the Request URL

```python
base_url = "https://api.wigle.net/api/v2/network/search"
```

```python
url_with_params = base_url + "?first=0&" + "latrange1="+str(latrange1)+"&latrange2="+str(latrange2)+"&longrange1="+str(longrange1)+"&longrange2="+str(longrange2)
```

```python
url_with_params
```

### Initiating GET Request

```python
response = requests.get(url_with_params,
            headers = {'Authorization': header})
```

### Response Status

```python
response
```

### Checking Response Error

```python
if response.json()['success']:
    print("Query Successful") 
else:
    print("Query Failed \nMessage: %s" %response.json()['message'])
```

```python
df = pd.DataFrame(response.json()["results"])
```

```python
df["color"] = np.where(df.encryption == "unknown","yellow",
                      np.where(df.encryption == "none","red","blue"))
```

### A Plain Map

```python
# create interactive map
map_osm = folium.Map(location=[latrange1, longrange2], zoom_start=10)
map_osm
```

### Map with desired Tile (Theme)

```python
folium.TileLayer('stamentoner').add_to(map_osm)
map_osm
```

### Map with Circle Markers of WiFi Spots

```python
#courtesy: https://stackoverflow.com/a/44475594/5086335
df.apply(lambda row:folium.CircleMarker(location=[row["trilat"], row["trilong"]], 
                                              radius=5, color=row['color'],
                                              popup = "SSID: %s /n Encrpytion %s" %(row["ssid"],row["encryption"]))
                                             .add_to(map_osm), axis=1)

map_osm
```
